```python
import pandas as pd
```

```python
df=pd.read_csv("tel_churn.csv")
df.head()
```

```python
df=df.drop('Unnamed: 0',axis=1)
df.info()
```

```python
df.describe()
```

```python
x=df.drop('Churn',axis=1)
x
```

```python
y=df['Churn']
y
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from imblearn.combine import SMOTEENN
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model_rf=RandomForestClassifier(n_estimators=100, criterion='gini', random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
sm = SMOTEENN()
X_resampled, y_resampled = sm.fit_resample(x, y)
```

```python
x.shape
```

```python
xr_train1,xr_test1,yr_train1,yr_test1=train_test_split(X_resampled, y_resampled,test_size=0.2)
```

```python
model_rf.fit(xr_train1,yr_train1)
```

```python
yr_predict1 = model_rf.predict(xr_test1)
```

```python
model_score_r1 = model_rf.score(xr_test1, yr_test1)
```

```python
print(model_score_r1)
print(metrics.classification_report(yr_test1, yr_predict1))
```

```python
print(metrics.confusion_matrix(yr_test1, yr_predict1))
```

```python
import pickle
filename = 'model.sav'
pickle.dump(model_rf, open(filename, 'wb'))
load_model = pickle.load(open(filename, 'rb'))
model_score_r1 = load_model.score(xr_test1, yr_test1)
model_score_r1
```
